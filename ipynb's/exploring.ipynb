### Groupby in Polars


```python
# Grab type 1 pokemon counts

print(pokemon.group_by("Type 1")
    .agg(
        pl.col("Name").count().alias("Type 1 count"),
         pl.col("Type 2").count().alias("Type 2 count")
         )
    .sort("Type 1 count",
          descending= True)
    )
```

```python
grouped = (pokemon.group_by("Generation")
        .agg(
            pl.mean("Speed"),
            pl.col("Type 2").count().alias("count"),
            pl.col("Name").count().alias("name_count"))
        .sort(
            "Generation", descending=False
            )
        )
print(grouped)
```

```python
gen1 = pokemon.filter(
    pl.col("Generation")==1
)
grouped = pokemon.group_by("Type 1").agg(
    pl.col("Name").count()
)
grouped.select(
    pl.col("Type 1"),
    pl.col("Name").sort(descending=False)
)
```
